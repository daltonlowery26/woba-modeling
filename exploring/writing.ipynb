```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

os.chdir('C:/Users/dalto/OneDrive/Pictures/Documents/Projects/Coding Projects/woba modeling/data/')
```

```python
pitch = pd.read_csv('pitch/pitch_cleaned.csv')
```

```python
pitch_ld = pitch[(pitch['launch_angle'] < 15) & (pitch['launch_angle'] > 12)  & (pitch['launch_speed'] < 98) & (pitch['launch_speed'] > 95)]
pitch_gb = pitch[(pitch['launch_angle'] < 3) & (pitch['launch_angle'] > 0)  & (pitch['launch_speed'] < 85) & (pitch['launch_speed'] > 82)]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.kdeplot(x=pitch_ld['woba_value'], fill=True, ax=axes[0])
axes[0].set_title("LA Between 12 and 15, EV Between 95 and 98")

sns.kdeplot(x=pitch_gb['woba_value'], fill=True, ax=axes[1])
axes[1].set_title("LA Between 0 and 3, EV Between 82 and 85")
plt.show()
```
